```python
import numpy as np
import matplotlib.pyplot as plt
loc = r'E:\OneDrive - TU Eindhoven\Master thesis\Measurements\Fiber_test'
def image_loc(x): return rf'C:\Users\20222772\PycharmProjects\MasterDataAnalysis\Images\Fiber_comparison\{x}'
```

```python
with open(rf"{loc}\Avantes_fiber.txt") as file:
    lines = file.readlines()
lines = [[float(l[0].replace(',', '.')), float(l[1].replace(',', '.'))] for l in [l.split(';') for l in lines[8:-2]]]
wav_av, sample_av = np.array(lines).T

with open(rf"{loc}\Ocean_fiber.txt") as file:
    lines = file.readlines()
lines = [[float(l[0].replace(',', '.')), float(l[1].replace(',', '.'))] for l in [l.split(';') for l in lines[8:-2]]]
wav_oc, sample_oc = np.array(lines).T

plt.figure()
plt.plot(wav_av, sample_av/np.max(sample_av), label='Avantes')
plt.plot(wav_oc, 1.125*sample_oc/np.max(sample_oc), label='Ocean')
plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative intensity (a.u.)')
plt.tight_layout()
plt.savefig(image_loc('fiber_comparison.png'))
plt.show()

plt.figure()
plt.plot(wav_av, sample_av/np.max(sample_av), label='Avantes')
plt.plot(wav_oc, sample_oc/np.max(sample_oc), label='Ocean')
plt.legend()
plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative intensity (a.u.)')
plt.xlim(190, 400)
plt.tight_layout()
plt.savefig(image_loc('fiber_comparison_zoomed.png'))
plt.show()

plt.figure()
plt.plot(wav_av, sample_av, label='Avantes')
plt.plot(wav_oc, sample_oc/1.25, label='Ocean')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Relative intensity (a.u.)')
# plt.xlim(190, 400)
plt.legend()

ax2 = plt.gca().twinx()
ax2.plot(wav_av, sample_av/sample_oc, color='red', label='Ratio')
plt.tight_layout()
plt.savefig(image_loc('fiber_comparison2.png'))
plt.show()
```
