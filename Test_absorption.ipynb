```python
import numpy as np
import matplotlib.pyplot as plt
from Import_data import import_txt
loc = r'E:\OneDrive - TU Eindhoven\Master thesis\Data\Test'
def image_loc(x): return rf'C:\Users\20222772\PycharmProjects\MasterDataAnalysis\Images\Test_absorption\{x}'
```

```python
files = ['cocktail1.txt', 'cocktail2.txt', 'cocktail3.txt', 'demi_end.txt']
labels = ['Cocktail 1', 'Cocktail 2', 'Cocktail 3', 'Demineralized end']
data = []
for file in files:
    data.append(import_txt(loc + '\\' + file))
```

```python
plt.figure()
for i in range(len(data)):
    plt.plot(data[i][0], -np.log10((data[i][1]-data[i][2])/(data[i][3]-data[i][2])), label=labels[i])
plt.legend()
plt.xlim(195, 420)
plt.ylabel('Absorption')
plt.xlabel('Wavelength (nm)')
plt.tight_layout()
plt.savefig(image_loc('absorption.png'))
plt.show()

plt.figure()
for i in range(len(data)):
    plt.semilogy(data[i][0], data[i][1]-data[i][2], label=labels[i])
plt.legend()
plt.xlim(195, 420)
plt.ylabel('Absorption')
plt.xlabel('Wavelength (nm)')
plt.tight_layout()
# plt.savefig(image_loc('absorption.png'))
plt.show()
```
